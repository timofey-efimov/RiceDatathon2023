```python
def process_images_ocrs(all_matches):
  all_processed = []
  count = 0
  for elem in all_matches:
    #print(count/len(all_matches)*100, " %")
    elemOCR = elem[0]
    elemIMG = elem[1]
    path_to_image = "/content/Bill.com/img/"
    path_to_ocr = "/content/Bill.com/ocr/"
    # print(path_to_image+elemOCR)
    # print(path_to_ocr+elemIMG)
    print(path_to_image+elemIMG)
    img = Image.open(path_to_image+elemIMG)
    # extract text
    text = pytesseract.image_to_string(img)
    # OCR
    # print(text)
    # print("=================")
    # print("=================")
    # print("=================")
    all_processed.append([text])
    #print(len(all_processed))
    # print(all_processed)
    count += 1
  return all_processed
```
