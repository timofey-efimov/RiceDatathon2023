```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import random 

# Create a string with the text you want to create a word cloud from
text = "This is a sample text that will be used to create a word cloud. The word cloud will show the most common words in the text and the size of the words will correspond to their frequency in the text."
colors = ["#FF0000", "#0000FF", "#00FF00", "#FFA500"]
def color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    return colors[random.randint(0, len(colors) - 1)]
# Create a word cloud object
wordcloud = WordCloud(color_func=color_func).generate(text)

# Display the word cloud
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Create a dictionary of words and their frequencies
words = {"word1": 10, "word2": 20, "word3": 30}

# Generate a word cloud with a white background
wc = WordCloud(background_color="white")
wc.generate_from_frequencies(words)

# Display the word cloud
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```
